### <u>Sets </u>
- unordered collection of unique objects # use-case is usernames, emails etc 

```
    $ my_set = {1,2,3,4,5}
    $ my_set.add(100)
    $ my_set.add(2) # does nothing
    $ my_list = [1,2,3,4,5,5]
    $ my_set = {1,2,3,4,5,5}
    $ set(my_list) 
    $ list(my_set)
```

#### <u> Set Methods </u>

```
    $ 5 in my_set
    $ len(my_set)
    $ # counts only unique values
    $ copy_set = my_set.copy() # complete new copy
```


#### <u> Frequently Used Methods  </u>

```
    $ set_one = {1,2,3,4,5}
    $ set_two = {4,5,6,7,8,9}
    $ set_one.difference(set_two)
    $ set_one.discard(2)
    $ set_one.difference_update(set_two) 
      # keeps only the different attributes
    $ set_one.intersection(set_two)
    $ set_one.isdisjoint(set_two)
      # returns TRUE or False 
    $ set_one.issubset(set_two)
    $ set_two.issuperset(set_one)
    $ set_one.union(set_two)
      # merges two sets, removes duplicates
    $ set_one | set_two
      # another declaration to merge without duplicates
 ```
